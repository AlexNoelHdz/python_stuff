```python
numero1 = 5
```

```python
numero2 = 3
```

```python
suma = numero1 + numero2
```

```python
strSuma = str(suma)
```

```python
type(strSuma)
```

```python
resultado = "El resultado es " + strSuma
resultado
```

```python
#string interpolation
resultado = f"El resultado es {strSuma}"
resultado
```

```python
#%-formatting
resultado = ("%s %s" %('El resultado es',strSuma,))
resultado
```

```python
print("%s %s" %('El resultado es ',strSuma,))
```

```python
#Str.format()
resultado = 'El resultado es {}'.format(strSuma)
resultado
```

```python
#template strings
from string import Template
texto = 'El resultado es '
new = Template('$texto $resultado.')
print(new.substitute(texto = texto, resultado = strSuma))
```
