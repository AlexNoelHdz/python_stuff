```python
def reverse_bits(n):
    numberInt = n
    numberBit = bin(numberInt)
    stringBit = str(numberBit)
    short_string_bit = stringBit[2:]
    reversed_string_bit = short_string_bit[::-1]
    return int(reversed_string_bit, 2)
```
